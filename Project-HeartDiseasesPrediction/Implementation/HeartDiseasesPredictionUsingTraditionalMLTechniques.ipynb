```python
def compareModels(arrayOfModels):
    modelsAndAccuracies = {}
    for data in enumerate(arrayOfModels):
        modelsAndAccuracies[data[1][4]]=data[1][3]
    bestModelAndItsAccuracy = {}
    bestModelAndItsAccuracy[max(modelsAndAccuracies.items(), key=operator.itemgetter(1))[0]]=modelsAndAccuracies[max(modelsAndAccuracies.items(), key=operator.itemgetter(1))[0]]

    sns.set_style("whitegrid")
    plt.figure(figsize=(10,30))
    plt.ylabel("Algorithms",fontsize=20)
    plt.xlabel("Accuracy %",fontsize=20)
    plt.title("Comparing the models based on the accuries achieved",fontsize=20)
    sns.barplot(x=list(modelsAndAccuracies.values()), y=list(modelsAndAccuracies.keys()))
    plt.show()
    return bestModelAndItsAccuracy
```

```python
def performPreprocessingBuildModelsAndEvaluateAccuracy(trainingDataSet,testingDataSet):
    for i in range(0,len(arrayOfModels)):
        print('***************************************************************************************************************************')
        print('********************************************* Building Model-', i ,' As Below *************************************************')
        print('\t -- Feature Encoding: \t ', arrayOfModels[i][0], ' \n\t -- Feature Scaling: \t ', arrayOfModels[i][1], ' \n\t -- Classification: \t ', arrayOfModels[i][2], '\n')
 
        trainingFileNameWithAbsolutePath = "D:\\Learning\\Deep Learning\\Project-HeartDiseasesPrediction\Datasets\\heart_train.csv"
        trainingDataSet = loadCSV(trainingFileNameWithAbsolutePath)
        labelName = getLabelName()
        label = trainingDataSet[labelName]

        testingFileNameWithAbsolutePath = "D:\\Learning\\Deep Learning\\Project-HeartDiseasesPrediction\Datasets\\heart_test.csv"
        testingDataSet = loadCSV(testingFileNameWithAbsolutePath)
        
        trainingDataSetAfterFeatuerSelection = trainingDataSet
        testingDataSetAfterFeatuerSelection = testingDataSet
        
        #Feature Encoding        
        if arrayOfModels[i][0] == 'LabelEncoder':
            #Perform lable encoding to convert categorical values into label encoded features
            trainingEncodedDataSet = featureEncodingUsingLabelEncoder(trainingDataSetAfterFeatuerSelection)
            testingEncodedDataSet = featureEncodingUsingLabelEncoder(testingDataSetAfterFeatuerSelection)
        elif arrayOfModels[i][0] == 'OneHotEncoder':
            #Perform OnHot encoding to convert categorical values into one-hot encoded features
            trainingEncodedDataSet = featureEncodingUsingOneHotEncoder(trainingDataSetAfterFeatuerSelection)
            testingEncodedDataSet = featureEncodingUsingOneHotEncoder(testingDataSetAfterFeatuerSelection)
        elif arrayOfModels[i][0] == 'FrequencyEncoder':
            #Perform Frequency encoding to convert categorical values into frequency encoded features
            trainingEncodedDataSet = featureEncodingUsingFrequencyEncoder(trainingDataSetAfterFeatuerSelection)
            testingEncodedDataSet = featureEncodingUsingFrequencyEncoder(testingDataSetAfterFeatuerSelection)
        elif arrayOfModels[i][0] == 'BinaryEncoder':
            #Perform Binary encoding to convert categorical values into binary encoded features
            trainingEncodedDataSet = featureEncodingUsingBinaryEncoder(trainingDataSetAfterFeatuerSelection)
            testingEncodedDataSet = featureEncodingUsingBinaryEncoder(testingDataSetAfterFeatuerSelection)

        #Feature Scaling        
        if arrayOfModels[i][1] == 'Min-Max':
            #Perform MinMaxScaler to scale the features of the dataset into same range
            trainingEncodedAndScaledDataset = featureScalingUsingMinMaxScaler(trainingEncodedDataSet)
            testingEncodedAndScaledDataset = featureScalingUsingMinMaxScaler(testingEncodedDataSet)
        elif arrayOfModels[i][1] == 'Binarizing':
            #Perform Binarizing to scale the features of the dataset into same range
            trainingEncodedAndScaledDataset = featureScalingUsingBinarizer(trainingEncodedDataSet)
            testingEncodedAndScaledDataset = featureScalingUsingBinarizer(testingEncodedDataSet)
        elif arrayOfModels[i][1] == 'Normalizing':
            #Perform Normalizing to scale the features of the dataset into same range
            trainingEncodedAndScaledDataset = featureScalingUsingNormalizer(trainingEncodedDataSet)
            testingEncodedAndScaledDataset = featureScalingUsingNormalizer(testingEncodedDataSet)
        elif arrayOfModels[i][1] == 'Standardization':
            #Perform Standardization to scale the features of the dataset into same range
            trainingEncodedAndScaledDataset = featureScalingUsingStandardScalar(trainingEncodedDataSet)
            testingEncodedAndScaledDataset = featureScalingUsingStandardScalar(testingEncodedDataSet)

        #Classification                
        if arrayOfModels[i][2] == 'DecisonTree':
            #Perform classification using DecisionTreeClassifier
            accuracyScore, classifier = classifyUsingDecisionTreeClassifier(trainingEncodedAndScaledDataset, testingEncodedAndScaledDataset)
        elif arrayOfModels[i][2] == 'RandomForestClassifier':
            accuracyScore, classifier = classifyUsingRandomForestClassifier(trainingEncodedAndScaledDataset, testingEncodedAndScaledDataset)
        elif arrayOfModels[i][2] == 'ExtraTreesClassifier':
            accuracyScore, classifier = classifyUsingExtraTreesClassifier(trainingEncodedAndScaledDataset, testingEncodedAndScaledDataset)
        elif arrayOfModels[i][2] == 'LogisticRegressionRegression':
            accuracyScore, classifier = classifyUsingLogisticRegression(trainingEncodedAndScaledDataset, testingEncodedAndScaledDataset)
        elif arrayOfModels[i][2] == 'LinearDiscriminantAnalysis':
            accuracyScore, classifier = classifyUsingLinearDiscriminantAnalysis(trainingEncodedAndScaledDataset, testingEncodedAndScaledDataset)
        elif arrayOfModels[i][2] == 'GuassianNaiveBayes':
            accuracyScore, classifier = classifyUsingGaussianNB(trainingEncodedAndScaledDataset, testingEncodedAndScaledDataset)

        arrayOfModels[i].append(accuracyScore)
        
        #Save the model to file
        modelName = arrayOfModels[i][0]+"_"+arrayOfModels[i][1]+"_"+arrayOfModels[i][2]
        modelFileName = modelName+".pkl"
        arrayOfModels[i].append(modelName)
        arrayOfModels[i].append(modelFileName)
        joblib.dump(classifier, modelFileName)
```
